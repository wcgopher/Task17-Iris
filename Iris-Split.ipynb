```python
import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
import sklearn.neighbors as nbrs
import sklearn.model_selection as model
import pandas as pd
%matplotlib notebook
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=['Sepal_Length', 'Sepal_Width','Petal_Length','Petal_Width'])
df['target']=iris.target
x = df.Petal_Length.values
y = df.Petal_Width.values
iris
```

```python
def create_colors():
    colors = []
    for i in range(150):
        if i <50:
            colors.append('r')
        elif i <100:
            colors.append('b')
        else:
            colors.append('k')
    return colors
colors = create_colors()
# colors
```

```python

ax1.scatter(df.Sepal_Length,df.Sepal_Width, c=colors)
ax2.scatter(df.Sepal_Length,df.Petal_Length, c=colors)
ax3.scatter(df.Sepal_Length,df.Petal_Width, c=colors)
ax4.scatter(df.Sepal_Width,df.Petal_Width, c=colors)
ax5.scatter(df.Sepal_Width,df.Petal_Length, c=colors)
ax6.scatter(df.Petal_Length,df.Petal_Width, c=colors)
```
